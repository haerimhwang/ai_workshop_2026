```python
## Cell 1

## Installing and loading packages

!pip install youtube-comment-downloader

import os
import json
import pandas as pd
os.chdir("/content/")
```

```python
## Cell 2

## Add your own Youtube link(s)
urls = [
    'https://youtu.be/3rkrvRlty5M?si=4ZAN81cfkE3NiI8Z',
    'https://youtu.be/3kRT-rsKxn4?si=-rGXYu784y2_mVt5'
]
```

```python
##  Cell 3

#   Initialize an empty list to store all information
all_data = []


#   Create a loop for each url
for i, url in enumerate(urls):

    # Name JSON
    json_file = f"YoutubeComments_{i}.json"

    # Use youtube-comment-downloader
    os.system(f"youtube-comment-downloader --url {url} --output {json_file}")

    # Read JSON
    with open(json_file, "r", encoding="utf-8") as f:
        json_data = []
        for line in f:
            json_data.append(json.loads(line))

    # JSON to DataFrame
    each_url_data = pd.DataFrame(json_data)

    # Add a column for the URL (to identify the source)
    each_url_data['video_url'] = url

    # Append the DataFrame to the list
    all_data.append(each_url_data)

    # Remove JSON
    os.remove(json_file)

#   Concatenate all DataFrames into one
combined_df = pd.concat(all_data, ignore_index=True)
```

```python
##  Cell 4

#   Save the combined DataFrame to a single CSV file
combined_df.to_csv("youtube_comments.csv", index=False, encoding="utf-8-sig")

#   See how "combined_df" looks
combined_df

```
