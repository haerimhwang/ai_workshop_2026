**2. Various ctext API demonstration calls 🙌**
*   Reference 👉  https://ctext.org/plugins/apilist/#gettextinfo

2.1. Returns a complete list of top-level textual items (typically books) in the textual database and Wiki.

```python
titles = gettexttitles()
print (titles)
```

2.2. Returns a list of available API functions and list of their parameters.

```python
capabilities = getcapabilities()
print (capabilities)
```

2.3. URL specifying a text on http://ctext.org.

```python
urn = readlink("https://ctext.org/xiyouji")
print (urn)
```

2.4. URN specifying the text to return.

```python
passages = gettext("ctp:analects/xue-er")
print(passages)
```

2.5. Returns the full text of the requested URN as an object simple list of strings, each corresponding to one chapter of text. Titles are omitted.

```python
chapters = gettextaschapterlist("ctp:analects")
print (chapters)
```

2.6. Returns the full text of the requested URN as an object with a nested structure representing what each gettext API call returns.

```python
data_1 = gettextasobject("ctp:analects/xue-er")
print (data_1)
```

2.7. Returns the full text of the requested URN as a simple list of strings, each corresponding to one passage of text. Titles are omitted.

```python
passages = gettextasparagraphlist("ctp:analects/xue-er")
print (passages)
```

2.8. Returns the full text of the requested URN as a single string. Each paragraph is separated with “\n\n”.



```python
string = gettextasstring("ctp:analects/xue-er")
print (string)
```

2.9. Simple wrapper around the gettextinfo API call.



```python
data_2 = gettextinfo("ctp:analects")
print(data_2)
```

3.1. Simple filter for lines containing “子曰” 👍

```python
xue_er = gettextaschapterlist("ctp:analects/xue-er")
for line in xue_er:
    if "子曰" in line:
        print(line)
        print()
```

```python
# @title 3.2. – DeepSeek API call counting “子曰” 🤓 {"display-mode":"form"}

# 1. OpenAI-compatible client setup
import os
import logging
from openai import OpenAI, OpenAIError
from google.colab import userdata

api_key = userdata.get("DEEPSEEK_API_KEY")

client = OpenAI(
    api_key=api_key,
    base_url="https://api.deepseek.com"
)

# 2. Ask a question and get the response

try:
    chat_response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"在{passages}中，'子曰'這詞一共出現多少次"}
        ]
    )

    print(chat_response.choices[0].message.content)

except OpenAIError as e:
    logging.error(f"An OpenAI API error occurred: {e}")
    if "502 Bad Gateway" in str(e):
        print("Error: 502 Bad Gateway. This usually indicates a temporary server issue with the DeepSeek API.")
        print("Please try running the cell again after a moment.")
    else:
        print(f"An unexpected API error occurred: {e}")
except Exception as e:
    logging.error(f"An unexpected error occurred: {e}")
    print(f"An unexpected error occurred: {e}")
```
